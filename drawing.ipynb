```python
import tkinter as tk

# Set the scale factor (how many physical pixels each logical pixel will be)
scale = 20
logical_size = 28
physical_size = logical_size * scale

# Create a master widget (window)
master = tk.Tk()

# Set the title of the window
master.title("Scaled 28x28 Canvas")

# Create a Canvas widget
canvas = tk.Canvas(master, bg='white', width=physical_size, height=physical_size)

# Pack the canvas into the master widget
canvas.pack()

# Function to draw on the canvas when the mouse is dragged
def paint(event):
    # Determine which logical pixel is being affected
    logical_x, logical_y = event.x / scale, event.y / scale

    # Calculate the top-left corner of the corresponding physical block
    x1, y1 = logical_x * scale, logical_y * scale
    x2, y2 = x1 + scale, y1 + scale

    # Draw a filled rectangle (block) for the logical pixel
    canvas.create_rectangle(x1, y1, x2, y2, fill='black', outline='')

# Bind the paint function to the mouse drag event
canvas.bind("<B1-Motion>", paint)

# Start the Tkinter event loop
master.mainloop()
```
